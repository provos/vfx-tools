```python
data = open('unreal.txt').read()
```

```python
lines = data.split('\n')
```

```python
import re

def getNumber(text):
    if text == '()':
        return 0.0
    text = text[1:-1]
    return float(text.split('=')[1])

def extract(data, relative=True):
    #print(data)
    times = re.match(r'.*Times=\((.*)\),Values=', data).group(1)
    values = re.match(r'.*Values=\((.*)\),(Default|TickResolution)', data).group(1)
    values=re.sub(r',Tangent=\([A-Za-z0-9-=,.]*\)', '', values)
    times = times.split(',')
    values = values.split(',')
    
    res = []
    for x, y in zip(times, values):
        res.append([int(getNumber(x)/1000.0), getNumber(y)])
        
    if relative:
        # make it relative to the first value
        for i in range(len(res) - 1):
            res[i+1][1] -= res[0][1]
        res[0][1] = 0
    
    return res
```

```python
translation = []
rotation = []

for line in lines:
    line = line.strip()
    if line.startswith('Translation'):
        data = extract(line)
        for i in range(len(data)):
            data[i][1] /= 100
        translation.append(data)
    if line.startswith('Rotation'):
        data = extract(line, relative=False)
        rotation.append(data)
```

__Swap Y and Z Axis__

```python
translation = [translation[0], translation[2], translation[1]]
rotation = [rotation[0], rotation[2], rotation[1]]
```

```python
print(translation)
print(rotation)
```

```python
import pickle

with open('translation.pkl', 'wb') as f:
  pickle.dump(translation, f)
  
with open('rotation.pkl', 'wb') as f:
  pickle.dump(rotation, f)
```
